```python
import pandas as pd
df = pd.read_csv("datasets/dataset_hw03.csv")
df.head()
```

```python
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

correlation_matrix = df.drop(columns='Income').corr()


X = df.drop(columns="Income") 
X_with_constant = sm.add_constant(X)  

vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X_with_constant.values, i+1) for i in range(len(X.columns))]
print("Correlation Matrix:")
print(correlation_matrix)

print("\nVIF Data:")
print(vif_data)
```

There is strong multicollinearity between Age and Years of Experience, which could affect the reliability of the regression model's coefficients

## Part 3: Build the Multiple Linear Regression Model

Now, I will build a multiple linear regression model using Income as the dependent variable and the other variables as independent predictors. I'll fit the model and display its statistical summary

```python
y = df['Income']  
X = df.drop(columns='Income') 
X_with_constant = sm.add_constant(X)
#Regression MOdel
model = sm.OLS(y, X_with_constant).fit()
model_summary = model.summary()
model_summary
```

```python
R-squared: 0.346
Adjusted R-squared: 0.343
F-statistic: 105.3 (p-value: 3.03e-89)
```
